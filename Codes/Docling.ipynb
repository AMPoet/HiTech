

# Convert .md to HTML




```python
from docling.document_converter import DocumentConverter # Import DocumentConverter
import os # Import os module

# Initialize converter
converter = DocumentConverter()

# input file
input_md = "your filename"
output_format = "html"

# Assuming 'ReadMe.md' exists in the current directory or a specified path
result = converter.convert(input_md)

# Access the converted document and export to HTML
document = result.document
html_content = document.export_to_html()

# Create output directory if it doesn't exist
output_dir = "your directory"
os.makedirs(output_dir, exist_ok=True)

# Get base filename without extension
base_filename = os.path.splitext(input_md)[0]

# Save HTML file
html_filename = os.path.join(output_dir, f"{base_filename}.html")
with open(html_filename, 'w', encoding='utf-8') as f:
    f.write(html_content)
print(f" HTML file saved in {html_filename}.")  # Show Output Path
```
